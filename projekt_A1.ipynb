## <div style="text-align: center"> Projekt

### <span style="color:red"> Proszę podać swoje imię, nazwisko i numer albumu: </span>

**Cel projektu:** Twoim zadaniem jest przeprowadzenie kompleksowej analizy historycznych cen akcji Apple. Wykorzystaj biblioteki do przetwarzania i wizualizacji danych. 
Analiza powinna uwzględniać różne aspekty związane z rynkiem akcji, w tym zmienność cen, wolumen transakcji i trendy czasowe.

***Interpretację wyników możesz dodawać w komentarzach w kodzie lub w oddzielnych komórkach Markdown.***

Opis kolumn:
- `Date` – Data notowania (YYYY-MM-DD)
- `Open` – Cena otwarcia danego dnia
- `High` – Najwyższa cena w ciągu dnia
- `Low` – Najniższa cena w ciągu dnia
- `Close` – Cena zamknięcia na koniec dnia
- `Volume` – Liczba akcji sprzedanych danego dnia
- `Adj Close` – Skorygowana cena zamknięcia (uwzględnia podziały akcji i dywidendy)
- `Year` – Rok (YYYY)
- `Weekday` – Wskazuje, w jaki dzień tygodnia miała miejsce dana sesja (np. Poniedziałek = 0, Wtorek = 1 itd.)
