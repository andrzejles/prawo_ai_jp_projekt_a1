Jaki był średni wolumen akcji Apple? Policz średnią, medianę oraz odchylenie standardowe i zinterpretuj wyniki dla kolumny `Volume`. Jaki była minimalny i maksymalny wolumen? 

</div>

```python
#df.Volume.describe()
#zwraca podstawowe statystyki, ale w niecyztelnej notacyji wykladniczej
#warto sformatowac wynik: 
#df.Volume.describe().apply('{:,.2f}'.format)
#zeby dostac tylko statystyki z polecenia:

print(f"Średnia:                {df['Volume'].mean():>20,.2f}")
print(f"Mediana:                {df['Volume'].median():>20,.2f}")
print(f"Odchylenie standardowe: {df['Volume'].std():>20,.2f}")

#Minimalny i maksymalny volumen
print(f"Minimum:                {df['Volume'].min():>20,.2f}")
print(f"Maksimum:               {df['Volume'].max():>20,.2f}")
```

**Interpretacja Zadania 8:**

1. **Średnia vs Mediana:** Średnia (~94 mln) jest znacznie wyższa od mediany (~66 mln), co wskazuje na **rozkład prawoskośny** - występują dni z ekstremalnie wysokim wolumenem, które zawyżają średnią.

2. **Odchylenie standardowe:** Wartość ~90 mln (prawie równa średniej!) świadczy o **bardzo dużej zmienności** wolumenu transakcji.

3. **Zakres:** Od 250 tys. do 1.85 mld akcji dziennie - różnica ponad 7000-krotna między minimalnym a maksymalnym wolumenem.

4. **Wniosek:** Wolumen transakcji Apple charakteryzuje się dużą zmiennością z dominacją wartości niższych i sporadycznymi dniami ekstremalnie wysokich obrotów (np. ważne ogłoszenia, premiery produktów).

**Interpretacja Zadania 13:**

1. **Efekt poniedziałkowy:** Poniedziałek ma **najniższy średni wolumen** (~87 mln), co potwierdza tzw. "efekt poniedziałkowy" - inwestorzy są ostrożniejsi na początku tygodnia po przerwie weekendowej.

2. **Środek tygodnia:** Środa ma **najwyższy wolumen** (~99 mln), podobnie czwartek i wtorek - to okres największej aktywności handlowej.
